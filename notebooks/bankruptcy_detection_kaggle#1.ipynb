```python
keys = [None]*96
for i in range(0,96):
    keys[i] = data_pd.columns[i]
    print(i)

keys
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
y = data_pd[keys[0]][:]
y
```

```python
x_dat = np.zeros((len(data_pd),95))
for i in range(1,96):
    x_dat[:,i-1] = data_pd[keys[i]][:]
    print(i)
x_dat
```

```python
x_dat.shape
```

```python
from sklearn.preprocessing import normalize
```

```python
for i in range(0,95):
    x_dat[:,i] = x_dat[:,i]/np.linalg.norm(x_dat[:,i])
x_dat
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_dat,y,test_size = 0.30, random_state = 42)
```

```python
X_train.shape
```

```python
model = keras.Sequential([
    layers.Dense(512,activation = 'relu', input_shape = [95]),
    layers.Dense(512,activation = 'relu'),
    layers.Dense(512,activation = 'relu'),
    layers.Dense(1,activation = 'sigmoid'),
])
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['binary_accuracy'],
)
```

```python
history = model.fit(
    X_train, y_train,
    batch_size = 256,
    epochs = 300
)
```

```python
history_df = pd.DataFrame(history.history)
history_df['loss'].plot()
```

```python
model.evaluate(X_test,y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
kfold = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 41)
```

```python
scores = []
```

```python
for train,test in kfold.split(x_dat,y):
    model = keras.Sequential([
    layers.Dense(512,activation = 'relu', input_shape = [95]),
    layers.Dense(512,activation = 'relu'),
    layers.Dense(512,activation = 'relu'),
    layers.Dense(1,activation = 'sigmoid'),
    ])
    model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['binary_accuracy'],
    )
    history = model.fit(
    x_dat[train], y[train],
    batch_size = 256,
    epochs = 300,
    verbose = 0,
    )
    t = model.evaluate(x_dat[test],y[test],verbose = 0)
    scores.append(t)
```

```python
scores
```
